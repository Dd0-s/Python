Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
import math

class ComplexNumber:
    def __init__(self, first, second, mode="alg"):
        if (not isinstance(first, (int, float))) or (not isinstance(second, (int, float))):
            raise ValueError("Components must be numbers")
        if mode == "alg":
            self.real = first
            self.imag = second
            self.mode = "alg"
        elif mode == "polar":
            self.real = first * math.cos(second)
            self.imag = first * math.sin(second)
            self.mode = "polar"
        else:
            raise Exception("mode only 'alg' or 'polar'")
        
    def __add__(self, other):
        if not isinstance(other, ComplexNumber):
            raise ValueError("Can only add a ComplexNumber")
        return ComplexNumber(self.real + other.real, self.imag + other.imag)
    
    def __sub__(self, other):
        if not isinstance(other, ComplexNumber):
            raise ValueError("Can only sub a ComplexNumber")
        return ComplexNumber(self.real - other.real, self.imag - other.imag)
    
    def __mul__(self, other):
        if not isinstance(other, ComplexNumber):
            raise ValueError("Can only mul a ComplexNumber")
        return ComplexNumber(self.real * other.real - self.imag * other.imag, \
                             self.real * other.imag + self.imag * other.real)
       
    def __truediv__(self, other):
        if not isinstance(other, ComplexNumber):
            raise ValueError("Can only mul a ComplexNumber")
        r = other.arg()
        if r == 0:
            raise ZeroDivisionError("Can only mul a non-zero ComplexNumber")
        return ComplexNumber((self.real * other.real + self.imag * other.imag) / r, \
                             -(self.real * other.imag + self.imag * other.real) / r)
                               
    def __str__(self):
        if self.mode == "alg":
            if self.imag != 0 and self.real == 0:
                return f"{round(self.imag, 2)}i"
            if self.imag > 0:
                return f"{round(self.real, 2)} + {round(self.imag, 2)}i"
            elif self.imag == 0:
                return f"{round(self.real, 2)}"
            else:
                return f"{round(self.real, 2)} - {round(abs(self.imag), 2)}i"
        else:
            return f"{round(self.modul(), 2)} * exp({round(self.arg(), 2)}i)"

    def polar(self):
        self.mode = "polar"
        return self.__str__()
            
    def alg(self):
        self.mode = "alg"
        return self.__str__()
        
    def modul(self):
        return math.sqrt(self.real**2 + self.imag**2)
    
    def arg(self):
        return math.atan2(self.imag, self.real)    
````

__Задание 3 (2 балла):__ Опишите класс для векторов в N-мерном пространстве. В качестве основы  используйте список значений координат вектора, задаваемый `list`. Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
class Vector:
    def __init__(self, components):
        if not isinstance(components, list):
            raise ValueError("Components must be a list")
        for component in components:
            if not isinstance(component, (int, float)):
                raise ValueError("Components must be numbers")
        self.components = components

    def __len__(self):
        return len(self.components)

    def __setitem__(self, index, value):
        if not isinstance(value, (int, float)):
            raise ValueError("Components must be numbers")
        self.components[index] = value
        
    def __getitem__(self, index):
        return self.components[index]

    def __add__(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Can only add a Vector")
        elif len(self) != len(other):
            raise ValueError("Vectors must be of same dimension")
        components = [self[i] + other[i] for i in range(len(self))]
        return Vector(components)

    def __sub__(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Can only sub a Vector")
        elif len(self) != len(other):
                raise ValueError("Vectors must be of same dimension")
        components = [self[i] - other[i] for i in range(len(self))]
        return Vector(components)

    def __mul__(self, other):
        if isinstance(other, Vector):
            if len(self) != len(other):
                raise ValueError("Vectors must be of same dimension")
            return sum([self[i] * other[i] for i in range(len(self))])
        elif isinstance(other, (int, float)):
            components = [other * i for i in self.components]
            return Vector(components)
        else:
            raise ValueError("Can only mul with a Vector or a scalar")
            
    def __rmul__(self, other):
        return self.__mul__(other)

    def __str__(self):
        return f"Vector({self.components})"        
    
    def cos(self, other):
        if not isinstance(other, Vector):
            raise ValueError("Can only cos a Vector")
        elif len(self) != len(other):
            raise ValueError("Vectors must be of same dimension")
        elif self.norm() == 0 or other.norm() == 0:
            raise ValueError("Cos is not defined for zero vectors")
        return self * other / (self.norm() * other.norm())

    def norm(self):
        return (sum([i**2 for i in self.components])) ** 0.5
````

__Задание 4 (2 балл):__ Опишите декоратор, который принимает на вход функцию и при каждом её вызове печатает строку "This function was called N times", где N - число раз, которое это функция была вызвана на текущий момент (пока функция существует как объект, это число, очевидно, может только неубывать).

````python
def calls_counter(func):
    
    def wrapper(*args, **kw):
        wrapper.count +=1
        print(f"This function was called {wrapper.count} times")
        return func(*args, **kw)
    
    wrapper.count = 0
    return wrapper
````

````python
#test
@calls_counter
def print_world():
    print("Hello, world")
    
@calls_counter
def print_world():
    print("Hello, world")
    
print_world()
print_world()
print_world()
print_world()
````

__Задание 5 (3 балла):__ Опишите декоратор класса, который принимает на вход другой класс и снабжает декорируемый класс всеми атрибутами входного класса, названия которых НЕ начинаются с "\_". В случае конфликтов имён импортируемый атрибут должен получить имя с суффиксом "\_new".

````python
def copy_class_attrs(cls):
    def wrapper(wrapped_cls):
        for attr in dir(cls):
            if not attr.startswith('_'):
                if hasattr(wrapped_cls, attr):
                    setattr(wrapped_cls, f"{attr}_new", getattr(cls, attr))
                else:
                    setattr(wrapped_cls, attr, getattr(cls, attr))
        return wrapped_cls
    return wrapper
````

````python
#test
class OtherClass:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self._c = c

@copy_class_attrs(OtherClass(1, 2, 3))
class MyClass:
    def __init__(self, x, y):
        self.x = x
        self.y = y

obj = MyClass(10, 20)
print(obj.a)
print(obj.b)
#print(obj.c) # Error
print(obj.x)
print(obj.y)
````

В реализации математических операций запрещается пользоваться любыми функциями, требующими использования оператора `import`.

````python
import pickle
from pathlib import Path
````

````python
class TextFile:
    def read_text(): pass  
    def save_text(): pass

class BinaryFile:
    def read_binary(): pass  
    def save_binary(): pass
    
class Copy:
    def copy(): pass
````

````python
class Matrix(TextFile, BinaryFile, Copy):
    def __init__(self, *args):
        if len(args) == 1:
            if isinstance(args[0], list):
                if isinstance(args[0][0], list):
                    for number_1 in args[0]:
                        if len(number_1) != len(args[0][0]):
                            raise ValueError("The length of all columns must be the same")
                        for number_2 in number_1:
                            if not isinstance(number_2, (int, float)):
                                raise ValueError("Components must be numbers")
                    self.matrix = args[0]
                else:
                    for number in args[0]:
                        if not isinstance(number, (int, float)):
                            raise ValueError("Components must be numbers")
                    self.matrix = [args[0]]
            else:
                raise ValueError("The matrix must be created from a list or a list of lists orexplicitly specified parameters")
        elif len(args) == 3:
            for number in args:
                if not isinstance(number, (int, float)):
                    raise ValueError("Components must be numbers")
            self.matrix = [[args[2]] * args[1] for _ in range(args[0])]
        else:
            raise ValueError("The matrix must be created from a list or explicitly specified parameters")
        self.rows = len(self.matrix)
        self.cols = len(self.matrix[0])

    def __str__(self):
        string = "["
        for i in range(self.rows):
            if i == self.rows - 1:
                string += str(self.matrix[i]) + "]" 
            else:
                string += str(self.matrix[i]) + "\n"
        return string

    def __getitem__(self, key):
        if 0 <= key < self.cols:
            return self.matrix[key]
        else:
            raise IndexError("The index beyond the boundaries of the matrix")
        return self.matrix[key]
    
    def __setitem__(self, index, value):
        if isinstance(index, tuple) and len(index) == 2:
            i, j = index
            if 0 <= i < self.rows and 0 <= j < self.cols:
                self.matrix[i][j] = value
            else:
                raise IndexError("The index beyond the boundaries of the matrix")
        else:
            raise TypeError("The index of the matrix should be given as a pair (i,j)")
                
    def row_iter(self):
        for row in self.matrix:
            yield row

    def col_iter(self):
        for col in range(self.cols):
            yield [self.matrix[row][col] for row in range(self.rows)]

    def __add__(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                result = Matrix(self.rows, self.cols, 0)
                for i in range(self.rows):
                    for j in range(self.cols):
                        result.matrix[i][j] = self.matrix[i][j] + other.matrix[i][j]
                return result
            else:
                raise ValueError("The sizes of the matrices do not match")
        else:
            raise TypeError("It is impossible to add a matrix with an object of another type")

    def __sub__(self, other):
        if isinstance(other, Matrix):
            if self.rows == other.rows and self.cols == other.cols:
                result = Matrix(self.rows, self.cols, 0)
                for i in range(self.rows):
                    for j in range(self.cols):
                        result.matrix[i][j] = self.matrix[i][j] - other.matrix[i][j]
                return result
            else:
                raise ValueError("The sizes of the matrices do not match")
        else:
            raise TypeError("It is impossible to sub a matrix with an object of another type")

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            result = Matrix(self.rows, self.cols, 0)
            for i in range(self.rows):
                for j in range(self.cols):
                    result.matrix[i][j] = self.matrix[i][j] * other
            return result
        elif isinstance(other, Matrix):
            if self.cols == other.rows:
                result = Matrix(self.rows, other.cols, 0)
                for i in range(self.rows):
                    for j in range(other.cols):
                        s = 0
                        for k in range(self.cols):
                            s += self.matrix[i][k] * other.matrix[k][j]
                        result[i][j] = s
                return result
            else:
                raise ValueError("Incompatible matrix sizes")
        else:
            raise TypeError("You can only multiply a matrix by a number or a matrix")
            
    def __rmul__(self, other):
        return self.__mul__(other)

    def dot(self, other):
        return self * other

    @property
    def T(self):
        result = Matrix(self.cols, self.rows, 0)
        for i in range(self.rows):
            for j in range(self.cols):
                result[j][i] = self[i][j]
        return result
    
    @property
    def tr(self):
        if self.rows == self.cols:
            return sum(self[i][i] for i in range(self.rows))
        else:
            raise ValueError("The matrix must be square")
            
    @property
    def det(self):
        if self.rows == self.cols:
            if self.rows == 1:
                return self.matrix[0][0]
            else:
                coef = 1
                result = 0
                for j in range(self.cols):
                    a = Matrix(self.rows - 1, self.cols - 1, 0)
                    for i in range(1, self.rows):
                        for k in range(self.cols):
                            if k < j:
                                a[i - 1][k] = self.matrix[i][k]
                            elif k > j:
                                a[i - 1][k - 1] = self.matrix[i][k]
                    result += coef * self.matrix[0][j] * a.det
                    coef = -coef
                return result
        else:
            raise ValueError("The matrix must be square")
            return None
  
    def read_text(path_to_file):
        with open(path_to_file, "r") as f: 
            matrix = []
            for line in f:
                row = []
                for coord in line.split():
                    row.append(float(coord))
                matrix.append(row)
        return Matrix(matrix)

    def save_text(self, path_to_file):
        with open(path_to_file, "w") as f:
            for i in range(self.rows):
                for j in range(self.cols):
                    if j == self.cols - 1:
                        f.write(f"{self.matrix[i][j]}\n")
                    else:
                        f.write(f"{self.matrix[i][j]} ")

    def read_binary(path_to_file):
        with open(path_to_file, 'rb') as f:
            matrix = pickle.load(f)
        return matrix                    
    
    def write_binary(self, path_to_file):
        with open(path_to_file, 'wb') as f:
            pickle.dump(self, f)
         
    def copy(self):
        matrix = self.matrix
        return Matrix(matrix)
````

````python
class Coffee:
    def __init__(self, coffee=None): 
        if coffee != None:
            raise Exception("You have to choose one coffee") 
    def calculate_cost(self):
        raise Exception("Choose the type of coffee")

    
class CoffeeDecorator(Coffee):
    
    def __init__(self, coffee=None): 
        if coffee == None:
            raise Exception("The first argument should be coffee") 
        self.coffee = coffee
                
    def calculate_cost(self):
        return self.coffee.calculate_cost()    

    
class Latte(Coffee):
    def calculate_cost(self):
        return 100

class Cappuccino(Coffee):
        def calculate_cost(self):
            return 200

class Americano(Coffee):
    def calculate_cost(self):
        return 300


class Sugar(CoffeeDecorator):
    def calculate_cost(self):
        return self.coffee.calculate_cost() + 10

class Cream(CoffeeDecorator):
    def calculate_cost(self):
        return self.coffee.calculate_cost() + 20

class Foam(CoffeeDecorator):
    def calculate_cost(self):
        return self.coffee.calculate_cost() + 30

class Syrup(CoffeeDecorator):
    def calculate_cost(self):
        return self.coffee.calculate_cost() + 40
````
